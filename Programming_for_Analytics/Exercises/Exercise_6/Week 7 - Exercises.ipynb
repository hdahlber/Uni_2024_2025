## Task 1

For the given NumPy array

    np.array([3,4,6,10,24,89,45,43,46,99,100])
    
please do the following:
    
 - Extract the numbers which are not divisible by 3
 - Extract the numbers which are divisible by 5
 - Set the numbers which are divisible by 3 to 0

```python
## Your program starts from here:

arr = np.array([3,4,6,10,24,89,45,43,46,99,100])

## Extract the numbers which are not divisible by 3
not_3 = arr[arr % 3 != 0]
print(not_3)



## Extract the numbers which are divisible by 5
div_5 = arr[arr% 5 == 0]
print(div_5)

## Set the numbers which are divisible by 3 to 0
numbers_to_0 = np.where(arr % 3 == 0, 0, arr)
print(numbers_to_0)
print(arr)
```

## Task 2  <code>frequency_of_head()</code>

Please write a function <code>frequency_of_head()</code> which has one argument, <code>no_of_toss</code>. The function should calculate and return the frequency of having a head after tossing a fair coin for a given number of times.  

**Hint**: You can use <code>np.random.randint()</code> to generate a sequence of random 0 and 1, where 1 represents a head and 0 represents a tail.

```python
## Your function starts from here:
def frequency_of_head(no_of_toss):
     return np.mean(np.random.randint(0, 2, size=no_of_toss) == 0)
```

```python
## Test your function with the following numbers (10, 100, and 100000):
##
## NOTE: the result given by your function should be different in each call.

print(frequency_of_head(10))

print(frequency_of_head(100))

print(frequency_of_head(100000))

## We could see that the result varies a lot in each call when no_of_toss is small (e.g. 10.)
## However, when the no_of_toss is fairly large (e.g. 100000), the result stays close to 0.50 in each call.
##
## This observation is called "The Law of Large Numbers".
##
## It states that, in statistics, as a sample size grows, their sample mean approaches their theoretical mean.
##
## This is not really required for or related to our course, but I thought you might be interesed in knowing why :)
```

## Task 3 <code>lottery_ticket()</code>

In order to win the top prize in a particular lottery, one must match all 6 numbers on his or her ticket to the 6 integers between 1 and 49 that are drawn by the lottery organizer. Please write a function <code>lottery_ticket()</code> that generates a random selection of 6 numbers for a lottery ticket. Please ensure that the 6 numbers selected do not contain any duplicates. Your function should return the numbers in ascending order.

```python
## Your function starts from here:
def lottery_ticket():
    return sorted(np.random.choice(range(1, 50), size=6, replace=False))
```

```python
## Call your function for a few times.
##
## NOTE: the numbers generated by your function should be different in each call.

for i in range(10):
    print(lottery_ticket())
```

## Task 1

Please recreate the following line plot for the Nominal GDP of Country A.



The Normal GDP of Country A between 1950 and 2010 is given as follows:

    gdp_a = np.array([300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3])

```python
## Your program starts from here:
gdp_a = np.array([300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3])
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
plt.plot(years,gdp_a, marker="o", color="blue")
plt.title("Nominal GDP of Contry A")
plt.xlabel("Year")
plt.ylabel("Billions of US dollars")

plt.show()

```

## Task 2

Please recreate the following line plot for the Nominal GDP of Country A, B and C.



The Normal GDP of Country A, B and C between 1950 and 2010 is given as follows:

    gdp_a = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
    gdp_b = [226.0, 362.0, 928.0, 1992.0, 4931.0, 7488.0, 12147.0]
    gdp_c = [1206.0, 1057.0, 1081.0, 2940.0, 8813.0, 13502.0, 19218.0]

```python
## Your program starts from here:
gdp_a = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
gdp_b = [226.0, 362.0, 928.0, 1992.0, 4931.0, 7488.0, 12147.0]
gdp_c = [1206.0, 1057.0, 1081.0, 2940.0, 8813.0, 13502.0, 19218.0]

plt.plot(years,gdp_a, marker="o", color="blue", label="Country A")
plt.plot(years, gdp_b, marker="*", color="red", linestyle=":", label="Country B")
plt.plot(years, gdp_c, marker="s", color="green",linestyle="-.", label="Country C")
plt.title("Nominal GDP")
plt.xlabel("Year")
plt.ylabel("Billions of US dollars")
plt.legend()
plt.show()
```
